```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"D:\Machine Learning\ML Algorithms\Datasets\insurance_data.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(data=dataset,x="age",y="bought_insurance")
plt.show()
```

```python
x = dataset[["age"]]
y = dataset['bought_insurance']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[21]])
```

```python
sns.scatterplot(data=dataset,x="age",y="bought_insurance")
sns.lineplot(data=dataset,x="age",y=lr.predict(x),c="red")
plt.show()
```

<H1>BINARY CLASSIFICATION / MULTIPLE INPUT<H1>

```python
df = pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Placement.csv")
df.head(3)
```

```python
sns.scatterplot(x="CGPA",y="IQ",data=df,hue="Placement")
```

```python
df.drop(columns=['Student_ID'],inplace=True)
```

```python
x = df.iloc[:,:-1]
y = df['Placement']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.score(X_test,Y_test)*100
```

```python
lr.predict([[8.14,130]])
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('/Users/shubhamyadav/Documents/D/Machine Learning/ML Algorithms/Datasets/Iris Raw.csv')
```

```python
dataset.drop(columns='Id',inplace=True)
dataset.head()
```

```python
sns.pairplot(data=dataset)
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
sc.dendrogram(sc.linkage(dataset,method='single',metric='euclidean')) #in dendogram linkage is asked there are diff methods and metrics shift +tab
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac = AgglomerativeClustering(n_clusters=2,linkage='single')
dataset['predict']=ac.fit_predict(dataset)
```

```python
sns.pairplot(data=dataset,hue='predict')
plt.show()
```

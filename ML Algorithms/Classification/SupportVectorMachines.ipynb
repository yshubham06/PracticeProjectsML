```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('/Users/shubhamyadav/Documents/D/Machine Learning/ML Algorithms/Datasets/placement.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(data= dataset,x = "cgpa",y="placement_exam_marks",hue="placed")
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel="linear") #{'linear', 'poly', 'rbf', 'sigmoid', 'precomputed'}
```

```python
svc.fit(x_train,y_train)
```

```python
svc.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=svc)
plt.show()
```

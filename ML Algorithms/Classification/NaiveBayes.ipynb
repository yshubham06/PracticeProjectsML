```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv('/Users/shubhamyadav/Documents/D/Machine Learning/ML Algorithms/Datasets/placement.csv')
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
sns.scatterplot(x="cgpa",y="placement_exam_marks",hue="placed",data=dataset)
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.2) 
```

```python
sns.kdeplot(data=dataset['cgpa'])
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
```

<h3>Naive Bayes - Gaussian</h3>

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
gnb.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

<h3>Naive Bayes - Multinomial</h3>

```python
mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
mnb.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

<h3>Naive Bayes - Bernoulli</h3>

```python
bnb = BernoulliNB()
```

```python
bnb.fit(x_train,y_train)
```

```python
bnb.score(x_test,y_test)
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=bnb)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('/Users/shubhamyadav/Documents/D/Machine Learning/ML Algorithms/Datasets/Social_Network_Ads.csv')
dataset.head()
```

```python
x = dataset.loc[:,["Age","EstimatedSalary"]]
y = dataset['Purchased']
```

```python
sns.scatterplot(data=dataset,x = "Age", y = "EstimatedSalary",hue= "Purchased")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(x)
```

```python
x = pd.DataFrame(ss.transform(x),columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
```

```python
KNN.fit(x_train,y_train)
```

```python
KNN.score(x_test,y_test)
```

```python
KNN.score(x_train,y_train)
```

```python
for i in range(1,30):
    knn1 = KNeighborsClassifier(n_neighbors=i)
    knn1.fit(x_train,y_train)
    print(i,knn1.score(x_train,y_train)*100,knn1.score(x_test,y_test)*100)
    
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=KNN)
plt.show()
```

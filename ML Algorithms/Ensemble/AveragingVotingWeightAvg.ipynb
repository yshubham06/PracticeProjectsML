```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
```

```python
x,y = make_moons(n_samples=1000,noise=0.2)
```

```python
df = {"x1":x[:,0],"x2":x[:,1],"y":y}
```

```python
dataset = pd.DataFrame(df)
```

```python
sns.scatterplot(data=dataset,x="x1",y="x2",hue="y")
```

```python
x_a = dataset.iloc[:,:-1]
y_a = dataset['y']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_a, y_a, random_state=42,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
dt.score(x_test,y_test)
```

```python
sv = SVC()
sv.fit(x_train,y_train)
sv.score(x_test,y_test)
```

```python
gnb = GaussianNB()
gnb.fit(x_train,y_train)
gnb.score(x_test,y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
li = [("dt1",DecisionTreeClassifier()),("sv1",SVC()),("gnb1",GaussianNB())]
```

```python
vc = VotingClassifier(li,weights=[10,5,7])
vc.fit(x_train,y_train)
vc.score(x_test,y_test)
```

```python
prd = {"dt":dt.predict(x_test), "svm":sv.predict(x_test), "gnb":gnb.predict(x_test),"vc":vc.predict(x_test)}
```

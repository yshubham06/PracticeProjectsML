```python
!pip install fitz
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io
```

```python
# Set up tesseract path if required
# pytesseract.pytesseract.tesseract_cmd = r'path_to_your_tesseract_executable'

def extract_images_from_pdf(pdf_path):
    # Open the PDF file
    pdf_file = fitz.open(pdf_path)
    
    images = []
    for page_num in range(len(pdf_file)):
        page = pdf_file.load_page(page_num)  # Get page
        image_list = page.get_images(full=True)  # Get all images from the page
        
        for img in image_list:
            xref = img[0]  # XREF of the image
            base_image = pdf_file.extract_image(xref)  # Extract the image bytes
            image_bytes = base_image["image"]
            images.append(image_bytes)  # Save image bytes for further processing

    return images

def extract_text_from_images(images):
    texts = []
    for img_bytes in images:
        image = Image.open(io.BytesIO(img_bytes))  # Open image from bytes
        text = pytesseract.image_to_string(image)  # Use tesseract to extract text
        texts.append(text)
    return texts

def extract_text_from_pdf_images(pdf_path):
    images = extract_images_from_pdf(pdf_path)  # Extract all images from the PDF
    texts = extract_text_from_images(images)  # Extract text from each image
    return "\n".join(texts)

```

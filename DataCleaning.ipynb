```python
# Forward elimination
```

```python
df_diabetes = pd.DataFrame(pd.read_csv("D:/Datasets/diabetes.csv"))
df_diabetes.head(3)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector
```

```python
x = df_diabetes.iloc[:,:-1]
y = df_diabetes['Outcome']
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
fs = SequentialFeatureSelector(lr,k_features = 5, forward= True)
```

```python
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_feature_names_
```

```python
fs.k_score_
```

```python
#backward elimination
```

```python
fs1 = SequentialFeatureSelector(lr,k_features = 5, forward= False)
```

```python
fs1.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_score_
```
